 –û—Å–Ω–æ–≤–Ω–æ–µ –Ω–∞–∑–Ω–∞—á–µ–Ω–∏–µ

–£–ø—Ä–æ—â–∞–µ—Ç –∫–æ–¥ (–Ω–µ –Ω—É–∂–Ω–æ –≤—Ä—É—á–Ω—É—é –≤—ã–∑—ã–≤–∞—Ç—å f.close()).

–ó–∞—â–∏—â–∞–µ—Ç –æ—Ç —É—Ç–µ—á–∫–∏ —Ä–µ—Å—É—Ä—Å–æ–≤ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ –ø—Ä–æ–≥—Ä–∞–º–º–∞ —É–ø–∞–¥—ë—Ç –ø—Ä–∏ —á—Ç–µ–Ω–∏–∏ —Ñ–∞–π–ª–∞).

–†–µ–∂–∏–º —Ä–∞–±–æ—Ç—ã (–æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ, –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 'r' ‚Äî —á—Ç–µ–Ω–∏–µ):

'+' ‚Äî —á—Ç–µ–Ω–∏–µ + –∑–∞–ø–∏—Å—å (–Ω–∞–ø—Ä–∏–º–µ—Ä, 'r+').

1. –ß—Ç–µ–Ω–∏–µ –∏ –≤—ã–≤–æ–¥ —Å–æ–¥–µ—Ä–∂–∏–º–æ–≥–æ —Ñ–∞–π–ª–∞
–ó–∞–¥–∞—á–∞:
–ü—Ä–æ—á–∏—Ç–∞–π—Ç–µ —Ñ–∞–π–ª poem.txt –∏ –≤—ã–≤–µ–¥–∏—Ç–µ –µ–≥–æ —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ –Ω–∞ —ç–∫—Ä–∞–Ω.

```python
with open('with1/poem.txt', 'r') as file:
    file = file.read()
    print(file)
```

 –ü–æ–¥—Å—á—ë—Ç —Å—Ç—Ä–æ–∫ –≤ —Ñ–∞–π–ª–µ

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –≤ —Ñ–∞–π–ª–µ data.txt.

```python
with open('with1/data.txt', 'r') as file:
    lines = file.readlines()
    print(f' –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –≤ —Ñ–∞–π–ª–µ "data.txt" = : {len(lines)}')
```

–ü–æ–∏—Å–∫ —Å–ª–æ–≤–∞ –≤ —Ñ–∞–π–ª–µ

```python
with open('with1/article.txt', 'r') as file:
    a = file.read()
    print('—Å–ª–æ–≤–æ –Ω–∞–π–¥–µ–Ω–æ!' if 'python' in a else '–º–ª–æ–≤–æ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç')
```

```python
with open('with1/article.txt', 'r') as file:
    a = file.read()
    if 'python' in a:
        print('—Å–ª–æ–≤–æ –Ω–∞–π–¥–µ–Ω–æ')
    else:
        print('—Ç–∞–∫–æ–≥–æ —Å–ª–æ–≤–∞ –Ω–µ—Ç')
```

–ó–∞–ø–∏—Å—å —Å–ø–∏—Å–∫–∞ –≤ —Ñ–∞–π–ª

```python
fruits = ['banana', 'mango', 'cherry']
with open('with1/fruits.txt', 'w') as file:
    for fruit in fruits:
        file.write(fruit + sep = '/n')
with open('with1/fruits.txt', 'r') as file:
    read = file.read()
    print(read)
```
